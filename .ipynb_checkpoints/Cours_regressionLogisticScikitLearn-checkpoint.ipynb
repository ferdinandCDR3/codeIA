```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
df = pd.read_excel('fruitDataset.xlsx')
df.head()
```

```python
nom_fruit_cible = dict( zip (df.etiquette_fruit.unique(), df.nom_fruit.unique()))
print( nom_fruit_cible )
#valeurs caractéristiques et valeur cible
x = df [['poids', 'largeur', 'hauteur']]
y = df['etiquette_fruit']
#fractionner dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)
#instanciation du modèle
modele_regLog = linear_model.LogisticRegression(random_state = 0, solver = 'liblinear', multi_class = 'ovr')
#training
modele_regLog.fit(x_train,y_train)
#précision du modèle
precision = modele_regLog.score(x_test,y_test)
print(precision*100)
```

```python
#prédiction
prediction_fruit = modele_regLog.predict([[180,8.0,6.8]])

## nom_fruit_cible[prediction_fruit[0]]
#prédiction
prediction_fruit = modele_regLog.predict([[20,4.3,5.5]])
print (prediction_fruit[0])
## nom_fruit_cible[prediction_fruit[0]]
```

```python
# Solution Exercice1
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model

df = pd.read_csv('iris.csv')
df.head()
a = df.loc[:,"PetalLengthCm"]
b = df.loc[:,"PetalWidthCm"]
x = list(zip(a, b))
y = df.loc[:,"Species"]
#fractionner dataset (train-test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
#instanciation du modèle
modele_regLog = linear_model.LogisticRegression(random_state = 0, solver = 'liblinear', multi_class = 'ovr')
#training
modele_regLog.fit(x_train, y_train)

precision = modele_regLog.score(x_train, y_train)
print("precision apprentissage \n")
print(precision * 100)
#calcule de précision
precision = modele_regLog.score(x_test, y_test)
# print(precision )
print("precision test \n")
print(precision * 100)
#Prédiction
longueur = 2.5
largeur = 0.75
prediction = modele_regLog.predict([[longueur, largeur]])
# print(prediction[0])
# affichage des résultats
resultat = "Résultat : " + prediction[0]
print(resultat)

# if 'setosa' in prediction[0]:
#     resultat = resultat + "setosa"
# if 'versicolor' in prediction[0]:
#     resultat = resultat + "versicolor"
# if 'virginica' in prediction[0]:
#     resultat = resultat + "virginica"
# print(resultat)
```

```python
# Solution Exercice2
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
nom_colonne = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi','pedigree', 'age', 'label']
df = pd.read_csv('diabetes.csv', header = None, names = nom_colonne)
df = df.drop([0])
df.head()
```

```python
caracteristique = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp', 'pedigree']
x = df[caracteristique]
y = df.label
#fractionner dataset (train-test)
x_train, x_test, y_train, y_test = train_test_split(x, y,
test_size = 0.10, random_state = 0)
#instanciation du modèle
modele_regLog = linear_model.LogisticRegression()
#training
modele_regLog.fit(x_train, y_train)

precision = modele_regLog.score(x_train, y_train)
print("precision apprentissage \n")
print(precision * 100)

#prediction
prediction = modele_regLog.predict(x_test)
print('prédiction du modèle:', prediction)
#calcule de précision
precision = modele_regLog.score(x_test, prediction)
print('la précision du modèle', precision * 100)

precision = modele_regLog.score(x_test, y_test)
# print(precision )
# print("precision test \n")
# print(precision * 100)
```
